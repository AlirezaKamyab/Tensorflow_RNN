# SunSpots Dataset

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import csv
import pandas as pd

# Configuring GPU device
physical_devices = tf.config.experimental.list_physical_devices('GPU')
tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

## Getting data ready

```python
df = pd.read_csv('archive/sunspot_data.csv')

# Remove missing values
df = df[df['Number of Sunspots'] != -1]

# Drop redundant values
df.drop(['Unnamed: 0'], axis=1, inplace=True)

# Not sure
df = df[df['Indicator'] != 0]
```

```python
df = df.groupby(['Year', 'Month']).mean()
```

```python
series = df['Number of Sunspots'].reset_index(drop=True)
```

```python
plt.figure(figsize=(10, 6))
plt.grid(True)
plt.plot(range(series.shape[0]), series);
plt.xlabel("Month")
plt.ylabel("Mean number of sunspots");
```

```python
time_steps = series.index.values
series = series.values
```

## Data split

```python
data_split = 2000
time_train = time_steps[:data_split]
x_train = series[:data_split]

time_val = time_steps[data_split:]
x_val = series[data_split:]
```

```python
def windowed_dataset(series, window_size, batch_size, shuffle_buffer_size=1000):
    dataset = tf.data.Dataset.from_tensor_slices(series)
    dataset = dataset.window(window_size + 1, shift=1, drop_remainder=True)
    dataset = dataset.flat_map(lambda w : w.batch(window_size + 1))
    dataset = dataset.map(lambda w : (w[:-1], w[-1]))
    dataset = dataset.shuffle(shuffle_buffer_size)
    dataset = dataset.batch(batch_size).prefetch(1)
    return dataset
```

```python
window_size = 125
batch_size = 128
shuffle_buffer_size = 1000
```

```python
dataset = windowed_dataset(x_train, window_size, batch_size, shuffle_buffer_size)
```

```python
def get_uncompiled_model(window_size):
    model = tf.keras.Sequential([
        tf.keras.layers.Conv1D(filters=50, kernel_size=3, strides=3, padding='causal', activation='relu',
                              input_shape=[window_size, 1]), 
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(50, return_sequences=True)),
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(50)),
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dense(1),
        tf.keras.layers.Lambda(lambda x: x * 100.0)
    ])
    return model
```

## Finding a good learning rate

```python
model = get_uncompiled_model(window_size)
model.summary()
```

```python
lr_schedule = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-8 * 10 ** (epoch / 20))
optimizer = tf.optimizers.SGD(momentum=0.9)
loss = tf.losses.Huber()
model.compile(optimizer=optimizer, loss=loss)

with tf.device('GPU:0'):
    history = model.fit(dataset, epochs=100, callbacks=[lr_schedule])
```

```python
plt.figure(figsize=(10, 6))
plt.grid(True)
lrs = 1e-8 * 10 ** (np.arange(100) / 20)
plt.semilogx(lrs, history.history['loss'])
plt.tick_params('both', length=10, width=1)

# choosen
learning_rate = 3e-6
plt.axvline(x=learning_rate, c='r', alpha=0.3);
```

## Train the model

```python
model = get_uncompiled_model(window_size)
optimizer = tf.optimizers.SGD(momentum=0.9, learning_rate=learning_rate)
loss = tf.losses.Huber()
model.compile(optimizer=optimizer, loss=loss, metrics=['mae'])

with tf.device('GPU:0'):
    history = model.fit(dataset, epochs=500)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 3))
ax[0].grid(True)
ax[0].plot(history.epoch, history.history['loss'])
ax[0].set_xlabel('epoch')
ax[0].set_ylabel('loss');

ax[1].grid(True)
ax[1].plot(history.epoch, history.history['loss'])
ax[1].set_xlim([300, 500])
ax[1].set_ylim([17, 20])
ax[1].set_xlabel('epoch')
ax[1].set_ylabel('loss');
```

## Prediction

```python
def model_predict(model, series, window_size, batch_size):
    dataset = tf.data.Dataset.from_tensor_slices(series)
    dataset = dataset.window(window_size, shift=1, drop_remainder=True)
    dataset = dataset.flat_map(lambda window: window.batch(window_size))
    dataset = dataset.batch(batch_size)
    return model.predict(dataset, verbose=0).squeeze()
```

```python
xs = series[data_split - window_size:-1]
y_hat = model_predict(model, xs, window_size, batch_size)

plt.figure(figsize=(10, 6))
plt.grid(True)
plt.plot(time_val, x_val, label='validation');
plt.plot(time_val, y_hat, label='prediction')
plt.legend();
plt.xlabel('Months');
plt.ylabel('Mean number of sunspots');
plt.title('Prediction and Validation plot');
```

```python
print(f'MSE is {tf.keras.metrics.MSE(x_val, y_hat)}')
print(f'MAE is {tf.keras.metrics.MAE(x_val, y_hat)}')
```
